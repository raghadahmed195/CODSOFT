```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier  
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
df=pd.read_csv('/kaggle/input/creditcardfraud/creditcard.csv')
df
```

```python
df.info()
```

```python
df['Class'].value_counts()
```

**data is inbalanced**

```python
X = df.drop(['Class'], axis=1)
y = df['Class']


# Standardize 'Amount' column
scaler = StandardScaler()
X['Amount'] = scaler.fit_transform(X[['Amount']])
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,random_state=42,test_size=.2,stratify=y)
```

```python
# Apply SMOTE to balance classes
smote = SMOTE(sampling_strategy=0.5, random_state=42)
xtrain_resampled, ytrain_resampled = smote.fit_resample(xtrain,ytrain)

# Check new class distribution
print(pd.Series(ytrain_resampled).value_counts())
```

# **logistic regression model**

```python

lr=LogisticRegression()
```

```python
lr.fit(xtrain_resampled,ytrain_resampled)
```

```python
lr.score(xtrain,ytrain)  #training accuracy
```

```python
lr.score(xtest,ytest)  #testing accuracy



```

```python
y_pred = lr.predict(xtest)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(ytest, y_pred))


print("\nClassification Report:")
print(classification_report(ytest, y_pred))
```

# **decision tree model** 

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(xtrain_resampled,ytrain_resampled)
```

```python
dt.score(xtrain,ytrain)  #training accuracy 
```

```python
dt.score(xtest,ytest) 
```

```python
y_pred = dt.predict(xtest)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(ytest, y_pred))

print("\nClassification Report:")
print(classification_report(ytest, y_pred))
```

# **Random forest**

```python
rf=RandomForestClassifier(n_estimators=25)
```

```python
rf.fit(xtrain_resampled,ytrain_resampled)
```

```python
rf.score(xtrain,ytrain)
```

```python

rf.score(xtest,ytest)
```
